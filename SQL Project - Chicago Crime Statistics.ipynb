```python
#List community area names and numbers with per capita income less than 11000
%sql SELECT COMMUNITY_AREA_NUMBER,COMMUNITY_AREA_NAME FROM CENSUS_DATA WHERE PER_CAPITA_INCOME < 11000
```

```python
#List all case numbers involving minors
%sql SELECT * from CHICAGO_CRIME_DATA WHERE DESCRIPTION LIKE "%MINOR%"
```

```python
#List all kidnapping crimes involving a child

%sql SELECT * FROM CHICAGO_CRIME_DATA WHERE PRIMARY_TYPE = "KIDNAPPING"
```

```python
#List all crimes recorded at schools

%sql SELECT DISTINCT(PRIMARY_TYPE) FROM CHICAGO_CRIME_DATA WHERE LOCATION_DESCRIPTION LIKE "%SCHOOL%"
```

```python
#List the Average Safety Scores of each type of school

%sql SELECT "Elementary, Middle, or High School", AVG(SAFETY_SCORE) AS AVERAGE_SAFETY_SCORE FROM CHICAGO_PUBLIC_SCHOOLS_DATA GROUP BY "Elementary, Middle, or High School";
```

```python
#List communities with highest percentage of households below the poverty line

%sql SELECT COMMUNITY_AREA_NAME FROM CENSUS_DATA ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY desc LIMIT 5
```

```python
#Find Community Area Number that has the highest crime frequency

%sql SELECT CCD.COMMUNITY_AREA_NUMBER ,COUNT(CCD.COMMUNITY_AREA_NUMBER) AS FREQUENCY FROM CHICAGO_CRIME_DATA AS CCD GROUP BY CCD.COMMUNITY_AREA_NUMBER ORDER BY COUNT(CCD.COMMUNITY_AREA_NUMBER) DESC LIMIT 1;
```

```python
#Create a sub-query finding the community area with the highest hardship index

%%sql

SELECT COMMUNITY_AREA_NAME
FROM  CENSUS_DATA
WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM CENSUS_DATA);
```

```python
#Create a sub-query determining the Community Area Name with the highest # of crimes

%%sql

SELECT COMMUNITY_AREA_NAME
FROM  CENSUS_DATA
WHERE HARDSHIP_INDEX = (SELECT MAX() FROM CENSUS_DATA);
```
